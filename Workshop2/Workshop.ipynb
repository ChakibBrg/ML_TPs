```python
import matplotlib
import numpy             as np
import pandas            as pd 
import matplotlib.pyplot as plt 
%matplotlib inline

np.__version__, pd.__version__, matplotlib.__version__
```

```python
sat_train = pd.read_csv('data/sat.trn', sep=' ', header=None)
sat_test  = pd.read_csv('data/sat.tst', sep=' ', header=None)

X_train = sat_train.iloc[:, :-1]
y_train = sat_train.iloc[:, -1]

X_test = sat_test.iloc[:, :-1]
y_test = sat_test.iloc[:, -1]

sat_train.head()
```

```python
plt.boxplot(X_train.iloc[:, :3])
plt.show()
```

```python
y_counts = y_train.value_counts().sort_index()
plt.pie(y_counts, labels=y_counts.index, autopct='%1.2f%%')
plt.show()
```

```python
from sklearn import datasets, decomposition

pca = decomposition.PCA(n_components=2)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca  = pca.transform(X_test)
```

```python
colors = ["blue", "orange", "green", "red", "purple", "brown"]
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```
